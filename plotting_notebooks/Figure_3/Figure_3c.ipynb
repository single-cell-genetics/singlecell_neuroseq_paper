```r
bulk_df = as.data.frame(t(t_df["ENSG00000171794",]))
colnames(bulk_df) = 'UTF1'
bulk_df$donor_id = rownames(bulk_df)
head(bulk_df)
```

```r
options(repr.plot.width=4, repr.plot.height=3.8)
library(dplyr)
library(ggplot2)
df = inner_join(diff_df, bulk_df, by = "donor_id")
nrow(df)
cor(df$diff_efficiency, df$UTF1)
cor.test(df$diff_efficiency, df$UTF1)
ggplot(df, aes(x = diff_efficiency, y = UTF1)) + geom_point(col='darkorange') + xlab("differentiation efficiency") +
    stat_smooth(method = "lm", se = F,col = "darkorange") + theme_bw() + theme(text = element_text(size=16)) 
```
